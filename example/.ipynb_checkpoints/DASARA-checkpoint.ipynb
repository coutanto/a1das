## ==> basic processing: spectrogram of a channel

### function: spectrogram

```python
def SpecPlot(a1, nfft=None,ch=0,fig=None, cmap='jet', vrange=None):
    #a1= DAS section
    #ch= channel number 
    
    '''
    PLot the spectrogram of a DAS channel
    ------------------------------------
    a1: A1Section class instance
    nfft: length of the FFT
    ch: channel id
    fig: segment duration (sec)
    cmap:colormap (default='jet')
    vrange: range for colormap
    
    '''
    from scipy import signal
    
    if not fig:
        fig,ax=plt.subplots(2,figsize=(12,7),constrained_layout=True)
        
        
    if not vrange:
        vmax = np.max(np.absolute(a1.data), axis=None)
        vmin = -vmax
    else:
        vmin = vrange[0]
        vmax = vrange[1]


    ax[0].plot(a1.time(),a1.data[:,ch],'k')
    ax[0].set_xlim(min(a1.time()),max(a1.time()))
    ax[0].set_ylabel("strain-rate [nm/m/s] ")

    if nfft is None:
        nperseg=256
    nperseg=nfft
    
    fs=1/a1.data_header['dt']
    freq, t, Sxx = signal.spectrogram(a1.data[:,ch],fs,nperseg=nperseg)
    s=ax[1].pcolormesh(t+min(a1.time()), freq, np.log10(Sxx/Sxx.max()), shading='gouraud',cmap=cmap,vmin=vmin,vmax=vmax)
    ax[1].set_xlabel('Time [sec]')
    ax[1].set_ylabel('Frequency [Hz]')
    ax[1].set_xlim(min(a1.time()),max(a1.time()))
    fig.colorbar(mappable=s, ax=ax[1])
    fig.suptitle('channel %d: %.2f m'%(ch,a1.dist()[ch]),fontsize=16)
```

### spectrogram at 3750 m 

les deux raies à 125Hz et 140 Hz sont potentiellement du bruit optique. 

```python
d=3753.6  #chosen distance
dist = a1.data_header['dist']  #istance along the fiber 
ch=np.nanargmin(np.abs(dist-d))  #channel number 

SpecPlot(a1, ch=ch,fig=None, cmap='jet', vrange=[0,-5])
```

### function: EnergyBand

```python
def EnergyBand(a1,fmin,fmax,dt,overlap=0.5,trange=None,drange=None):
    '''
    Waterfall View (~ EnergyBand) of a DAS section as an image
    ------------------------------------
    a1: A1Section class instance
    fmin: min. frequency (Hz)
    fmax: max. frequency (Hz)
    dt: segment duration (sec)
    overlap: overlapping percentage
    trange: time range (m)
    drange: distance range(m)
    '''
    from scipy.fftpack import fft
    
    
    #time & distance range: indices
    
    dist = a1.data_header['dist']
    time = a1.data_header['time']
    
    if drange is None:
        d1=0
        if transpose:
            d2 = a1.data.shape[1]
        else:
            d2=a1.data.shape[0]
        dmin=dist[0]
        dmax=dist[-1]
    else:
        dmin=drange[0]
        dmax=drange[1]
        d1=np.nanargmin(np.abs(dist-dmin))
        d2=np.nanargmin(np.abs(dist-dmax))

    if trange is None:
        t1=0
        if transpose:
            t2 = a1.data.shape[1]
        else:
            t2 = a1.data.shape[0]
        tmin=time[0]
        tmax=time[-1]
    else:
        tmin=trange[0]
        tmax=trange[1]
        t1=np.nanargmin(np.abs(time-tmin))
        t2=np.nanargmin(np.abs(time-tmax))
    
    #parameters 
    
    fs=1/a1.data_header['dt']
    dx=a1.data_header['dx']
    
    nfft=int(dt*fs)
    noverlap=int(np.floor(nfft * overlap))
    
    #time & distance range: extraction
    time2=time[t1:t2 - nfft:noverlap]
    ntime=len(time2)
    
    dist2=dist[d1:d2]
    nspace=len(dist2)
    
    nwin=int(ntime/nfft)
    
    freq=np.linspace(0,fs,nfft) 
    
    #print(fs,dx,nfft,noverlap,ntime,nspace)
    #print(time2)
    
    E=np.zeros((ntime,nspace))
    for time_id in range(ntime):
        start = time_id * noverlap
        end = start + nfft
        strain=a1.data[start:end,d1:d2]
        magnitude=abs(fft(strain,axis=0))
        energy=np.sum(magnitude[np.where((freq>fmin)&(freq<fmax)),:],axis=1)
        E[time_id]=energy
    return time2, dist2, E
    
    
```

### function: EnergyBandPlot

```python
def EnergyBandPlot(time,distance,energy,figsize=None,vrange=None, cmap='jet',title=''):
    '''
    EnergyBandPlot
    ------------------------------------
    time: time id [sec]
    distance: distance id[m]
    energy: energy [nt,ns]
    figsize: figsize(sec)
    overlap: overlapping percentage
    vrange: range for colormap
    cmap:colormap (default='jet')
    title: 
    ''' 
    if not figsize:
        fig,ax = plt.subplots(figsize=(12, 7))
        
    fig,ax = plt.subplots(figsize=figsize)
    
    if not vrange:
        vmax = np.max(energy, axis=None)
        vmin = 0
    else:
        vmin = vrange[0]
        vmax = vrange[1]

        
    dmin, dmax=distance[0],distance[-1]
    tmin, tmax=time[0],time[-1]
    extent=[dmin,dmax,tmax,tmin]
    
    pos=ax.imshow(energy,cmap=cmap,aspect='auto',vmin=vmin, vmax=vmax, extent=extent)

    
    ax.set_xlabel('distance along the fiber [meters]')
    ax.set_ylabel('time [sec]')
    ax.set_title('Waterfall View ' + title)
    fig.colorbar(pos)
```

### waterfall wiew

On choisit des fenêtres de 0.5s qui s'overlappent de 50% et on regarde entre 25 et 100Hz. 


Les tirs apparaissent clairement et le maximum d'énergie associé est spatialement identifiable (~ 90m). 

```python

tmin,tmax=132,175  #sec
dmin,dmax=30,200    #meters

time,distance,energy=EnergyBand(a1,fmin=25,fmax=100,dt=.5,overlap=0.5,trange=[tmin,tmax],drange=[dmin,dmax])
EnergyBandPlot(time,distance,energy,figsize=(9, 9),cmap='hot')
```

### position du capteur virtuel= 86.4 m

En analysant les courbes d'énergie on identifie la position de la chambre de tirage à 86.4 m. Il faut ensuite géoreferencer cette position!


```python
tmin,tmax=152,154.5  #sec

t1=np.nanargmin(np.abs(time-tmin))
t2=np.nanargmin(np.abs(time-tmax))



plt.figure()
plt.plot(distance,energy[t1:t2,:].T)
plt.xlabel('distance [meters]')
plt.ylabel('Energy')


val=np.array([time[t1:t2],distance[np.argmax(energy[t1:t2,:],axis=1)],np.max(energy[t1:t2,:],axis=1)])

for i in range(val.shape[1]):
    print('[%.3f s] => %.3f m: %.2f'%(val[0,i],val[1,i],val[2,i]))
```
